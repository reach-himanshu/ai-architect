# Design Patterns 04: Structural

Adapter, Decorator, and Facade.

## 1. Adapter
Making the "Old" Interface look like the "New" one.

```python
class OldSystem:
    def legacy_request(self):
        return "Legacy Output"

class Adapter:
    def __init__(self, old_system):
        self.old_system = old_system

    def standard_request(self):
        # Translate/Adapt the response
        return f"Adapted: {self.old_system.legacy_request()}"

adapter = Adapter(OldSystem())
print(adapter.standard_request())
```

## 2. Facade
Hiding complexity behind a simple button.

```python
class CPU:
    def freeze(self): print("CPU Freezing")
class Memory:
    def load(self): print("Memory Loading")
class HardDrive:
    def read(self): print("HD Reading")

class ComputerFacade:
    def __init__(self):
        self.cpu = CPU()
        self.mem = Memory()
        self.hd = HardDrive()

    def start(self):
        print("--- Starting Computer ---")
        self.cpu.freeze()
        self.mem.load()
        self.hd.read()

ComputerFacade().start()
```
