# Design Patterns 02: Dependency Injection (DI)

Decoupling classes for better testing and flexibility.

## 1. The "Before" (Tight Coupling)
Here, `OrderService` creates its own `EmailSender`. We cannot test `OrderService` without sending real emails!

```python
class EmailSender:
    def send(self, msg):
        print(f"SENDING REAL EMAIL TO SERVER: {msg}")

class OrderService:
    def __init__(self):
        self.emailer = EmailSender() # <--- Hardcoded dependency

    def place_order(self, order_id):
        self.emailer.send(f"Order {order_id} placed!")

# OrderService().place_order(101)
```

## 2. The "After" (Dependency Injection)
We ask for the emailer in the constructor. Now we can swap it.

```python
class OrderServiceDI:
    def __init__(self, emailer): # <--- Injected dependency
        self.emailer = emailer

    def place_order(self, order_id):
        self.emailer.send(f"Order {order_id} placed!")

class MockEmailSender:
    def send(self, msg):
        print(f"(TESTING) Fake email recorded: {msg}")

# Production Use
real_service = OrderServiceDI(EmailSender())
real_service.place_order(202)

# Test Use
test_service = OrderServiceDI(MockEmailSender())
test_service.place_order(202)
```
