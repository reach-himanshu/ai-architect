# Design Patterns 07: Repository & Registry

Decoupling data access and managing global services.

## 1. Repository Pattern
Hiding the database details.

```python
from abc import ABC, abstractmethod

class UserRepository(ABC):
    @abstractmethod
    def get_by_id(self, user_id):
        pass

class SqlRepository(UserRepository):
    def get_by_id(self, user_id):
        return f"SELECT * FROM users WHERE id={user_id}"

class InMemoryRepository(UserRepository):
    def get_by_id(self, user_id):
        return {"id": user_id, "name": "Test User"}

repos = [SqlRepository(), InMemoryRepository()]
for repo in repos:
    print(repo.get_by_id(1))
```

## 2. Registry Pattern
A simple service locator.

```python
class ServiceRegistry:
    _services = {}

    @classmethod
    def register(cls, name, service):
        cls._services[name] = service

    @classmethod
    def get(cls, name):
        return cls._services.get(name)

ServiceRegistry.register("db", SqlRepository())
print(ServiceRegistry.get("db").get_by_id(50))
```
