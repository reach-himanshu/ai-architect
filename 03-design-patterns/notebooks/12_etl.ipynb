# Design Patterns 12: ETL

Building a functional data pipeline.

```python
class Pipeline:
    def __init__(self, source):
        self.data = source

    def pipe(self, func):
        self.data = func(self.data)
        return self

    def collect(self):
        return self.data

# 1. EXTRACT
def extract_data(dummy_input):
    print("Extracting...")
    return [
        {"name": "Alice", "age": 25, "active": True},
        {"name": "Bob", "age": 17, "active": False},
        {"name": "Charlie", "age": 35, "active": True}
    ]

# 2. TRANSFORM
def filter_active(data):
    return [x for x in data if x['active']]

def uppercase_names(data):
    for x in data:
        x['name'] = x['name'].upper()
    return data

# 3. LOAD
def load_data(data):
    print("Loading to target...")
    for record in data:
        print(f"SAVED: {record}")
    return data

# Execute Pipeline
Pipeline(None) \
    .pipe(extract_data) \
    .pipe(filter_active) \
    .pipe(uppercase_names) \
    .pipe(load_data)
```
