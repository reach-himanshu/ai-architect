# Design Patterns 15: Batching & Paging

Simulating Offset vs Cursor pagination.

```python
# Generating dummy data [0, 1, 2 ... 99]
db_data = list(range(100))

def get_page_offset(page_num, page_size):
    start = (page_num - 1) * page_size
    end = start + page_size
    # Slice represents OFFSET + LIMIT
    return db_data[start:end]

def get_page_cursor(last_seen_id, page_size):
    # Simulating WHERE id > last_seen_id
    results = []
    count = 0
    for item in db_data:
        if item > last_seen_id:
            results.append(item)
            count += 1
            if count == page_size:
                break
    return results

print(f"Offset Page 2 (size 5): {get_page_offset(2, 5)}")
print(f"Cursor Next 5 after ID 4: {get_page_cursor(4, 5)}")
```
