# Design Patterns 16: Streaming

Using Generators to process infinite data.

```python
import sys

# 1. Batch Approach (Memory Heavy)
def batch_range(n):
    return [i for i in range(n)]

# 2. Stream Approach (Memory Efficient)
def stream_range(n):
    for i in range(n):
        yield i

limit = 1_000_000
print(f"Batch Size (List): {sys.getsizeof(batch_range(limit))} bytes")
print(f"Stream Size (Gen): {sys.getsizeof(stream_range(limit))} bytes")
```
