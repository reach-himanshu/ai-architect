# Design Patterns 13: Microservices (API Gateway)

Simulating a gateway routing requests to services.

```python
# Service 1
class UserService:
    def get_user(self, uid): return f"Data for User {uid}"

# Service 2
class OrderService:
    def get_order(self, oid): return f"Details for Order {oid}"

# API Gateway (The Pattern)
class ApiGateway:
    def __init__(self):
        self.user_service = UserService()
        self.order_service = OrderService()

    def route_request(self, path, param):
        print(f"Gateway Routing: {path}/{param}")
        if path == "/users":
            return self.user_service.get_user(param)
        elif path == "/orders":
            return self.order_service.get_order(param)
        else:
            return "404 Not Found"

gateway = ApiGateway()
print(gateway.route_request("/users", 101))
print(gateway.route_request("/orders", 500))
```
