# Design Patterns 14: Caching

Implementing the **Cache-Aside** pattern.

```python
class Database:
    def get(self, key):
        print(f"[DB] Fetching {key}...")
        return f"Data for {key}"

class Cache:
    def __init__(self):
        self.store = {}

    def get(self, key):
        return self.store.get(key)

    def set(self, key, value):
        print(f"[Cache] Storing {key}...")
        self.store[key] = value

class App:
    def __init__(self):
        self.db = Database()
        self.cache = Cache()

    def get_data(self, key):
        # 1. Check Cache
        val = self.cache.get(key)
        if val:
            print(f"[App] Cache Hit: {val}")
            return val

        # 2. Check DB (Cache Miss)
        print("[App] Cache Miss.")
        val = self.db.get(key)

        # 3. Update Cache
        self.cache.set(key, val)
        return val

app = App()
app.get_data("user_1") # Miss
app.get_data("user_1") # Hit
```
