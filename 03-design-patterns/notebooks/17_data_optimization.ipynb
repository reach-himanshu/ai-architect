# Design Patterns 17: Data Optimization

Simulating size savings of binary packing vs JSON.

```python
import json
import struct
import sys

data = {"id": 1, "val": 500}

# 1. JSON
json_data = json.dumps(data)
print(f"JSON Size: {len(json_data.encode('utf-8'))} bytes")

# 2. Binary Packing (Struct)
# 'i' = integer (4 bytes), 'i' = integer (4 bytes)
packed_data = struct.pack('ii', data['id'], data['val'])
print(f"Binary Size: {len(packed_data)} bytes")
print(f"Unpacked: {struct.unpack('ii', packed_data)}")
```
