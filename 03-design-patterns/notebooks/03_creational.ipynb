# Design Patterns 03: Creational

Singleton, Factory, and Builder.

## 1. Singleton
Using `__new__` to control instance creation.

```python
class Singleton:
    _instance = None
    
    def __new__(cls, *args, **kwargs):
        if not cls._instance:
            print("Creating the single instance")
            cls._instance = super().__new__(cls)
        return cls._instance

s1 = Singleton()
s2 = Singleton()

print(f"Are they the same object? {s1 is s2}")
```

## 2. Factory Method
Creating logic without knowing the exact class.

```python
class Dog:
    def speak(self): return "Woof!"
class Cat:
    def speak(self): return "Meow!"

class PetFactory:
    def get_pet(self, pet_type):
        pets = dict(dog=Dog(), cat=Cat())
        return pets.get(pet_type, None)

factory = PetFactory()
print(factory.get_pet("cat").speak())
```

## 3. Builder Pattern
Chaining methods to build an object.

```python
class QueryBuilder:
    def __init__(self):
        self.table = ""
        self.fields = []

    def select(self, *args):
        self.fields = list(args)
        return self # Return self for chaining

    def from_table(self, table):
        self.table = table
        return self

    def build(self):
        return f"SELECT {', '.join(self.fields)} FROM {self.table}"

query = QueryBuilder().select("id", "name").from_table("users").build()
print(query)
```
