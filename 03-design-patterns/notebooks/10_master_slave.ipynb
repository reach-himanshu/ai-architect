# Design Patterns 10: Master-Slave Replication

Simulating a database cluster with read/write splitting.

```python
import random

class Database:
    def __init__(self, name, check_roles=True):
        self.name = name
        self.data = {}
        self.check_roles = check_roles

class Master(Database):
    def __init__(self, slaves):
        super().__init__("Master")
        self.slaves = slaves

    def write(self, key, value):
        print(f"[Master] Writing {key}={value}")
        self.data[key] = value
        self.replicate(key, value)

    def replicate(self, key, value):
        for slave in self.slaves:
            slave.sync(key, value)

class Slave(Database):
    def sync(self, key, value):
        print(f"[{self.name}] Replicating {key}...")
        self.data[key] = value

    def read(self, key):
        return self.data.get(key, None)

# Setup Cluster
slave1 = Slave("Slave 1")
slave2 = Slave("Slave 2")
master = Master([slave1, slave2])

# App Logic
master.write("user_1", "Alice")

# Read from random slave
reader = random.choice([slave1, slave2])
print(f"Reading form {reader.name}: {reader.read('user_1')}")
```
