# Design Patterns 06: MVC Architecture

Simulating a simple MVC flow.

```python
# 1. MODEL: Data and Business Logic
class User:
    def __init__(self, name, age):
        self.name = name
        self.age = age

# 2. VIEW: Display Logic
class UserView:
    def render(self, user):
        print(f"[HTML] <h1>Profile: {user.name} ({user.age})</h1>")

# 3. CONTROLLER: Coordination
class UserController:
    def __init__(self, view):
        self.view = view

    def get_user(self, user_id):
        # Simulating DB fetch
        user = User("Himanshu", 30)
        self.view.render(user)

controller = UserController(UserView())
controller.get_user(101)
```
