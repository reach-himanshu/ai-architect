# Python Memory Model

References, mutability, and copying.

```python
# 1. REFERENCES
a = [1, 2, 3]
b = a  # Same object

print(f"id(a): {id(a)}")
print(f"id(b): {id(b)}")
print(f"a is b: {a is b}")
```

```python
# 2. SHALLOW VS DEEP COPY
import copy

original = [[1, 2], [3, 4]]
shallow = copy.copy(original)
deep = copy.deepcopy(original)

original[0].append(99)

print(f"Original: {original}")
print(f"Shallow: {shallow}")  # Affected!
print(f"Deep: {deep}")        # Not affected
```

```python
# 3. MUTABLE DEFAULT ARGUMENT PITFALL
def bad_append(item, lst=[]):
    lst.append(item)
    return lst

print(bad_append(1))  # [1]
print(bad_append(2))  # [1, 2] - Same list reused!
```
