# Python Loops

Loops are used to repeat a block of code multiple times. Python has two primitive loop commands: `for` and `while`.

## 1. The `for` Loop
Iterating over a sequence (list, tuple, string, etc.).

```python
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
for fruit in fruits:
    print(f"I like {fruit}")

print("\n")
for f in fruits:
    if len(f) > 5:
        print(f"{f} is a long-named fruit")
    if 'y' in f.lower():
        print(f"{f} contains the letter 'y'")
print("\n")
```

### Using `range()`
The `range()` function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default).

```python
print("Looping from 0 to 4:")
for i in range(5):
    print(i)
```

## 2. The `while` Loop
Executes as long as a condition is true.

```python
count = 0
while count < 3:
    print(f"Count is {count}")
    count += 1
```

## 3. Loop Control: `break` and `continue`

```python
print("Breaking at 3:")
for i in range(6):
    if i == 3:
        break
    print(i)

print("\nSkipping 3:")
for i in range(6):
    if i == 3:
        continue
    print(i)
```

## 4. Loop with `else`
The `else` block runs when the loop finishes its iteration without being interrupted by a `break` statement.

```python
for i in range(3):
    print(i)
else:
    print("Loop completed naturally.")
```

## 5. The `pass` Statement
The `pass` statement is used as a placeholder. It does nothing and is used when a statement is syntactically required but no action is needed.

```python
for x in [0, 1, 2]:
    if x == 1:
        pass  # Placeholder for future logic
        
    else:
        print(x)
```
