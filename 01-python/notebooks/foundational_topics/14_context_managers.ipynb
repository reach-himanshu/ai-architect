# Context Managers

Guaranteed resource cleanup with `with`.

```python
# 1. CLASS-BASED CONTEXT MANAGER
class Timer:
    import time
    def __enter__(self):
        self.start = Timer.time.time()
        return self

    def __exit__(self, *args):
        self.elapsed = Timer.time.time() - self.start
        print(f"Elapsed: {self.elapsed:.4f}s")

with Timer():
    sum(range(1_000_000))
```

```python
# 2. GENERATOR-BASED CONTEXT MANAGER
from contextlib import contextmanager

@contextmanager
def temp_change_setting(obj, attr, new_value):
    old_value = getattr(obj, attr)
    setattr(obj, attr, new_value)
    try:
        yield
    finally:
        setattr(obj, attr, old_value)

class Config:
    debug = False

print(f"Before: {Config.debug}")
with temp_change_setting(Config, 'debug', True):
    print(f"Inside: {Config.debug}")
print(f"After: {Config.debug}")
```
