# Async/Await & Asyncio

Concurrent programming for I/O-bound tasks.

```python
import asyncio

async def fetch(url, delay):
    print(f"Fetching {url}...")
    await asyncio.sleep(delay)  # Simulating network I/O
    return f"Data from {url}"

async def main():
    # Run concurrently - total time = max(delays), not sum
    results = await asyncio.gather(
        fetch("api.com/users", 2),
        fetch("api.com/orders", 1),
        fetch("api.com/products", 3)
    )
    for r in results:
        print(r)

await main()  # In Jupyter; use asyncio.run(main()) in scripts
```
