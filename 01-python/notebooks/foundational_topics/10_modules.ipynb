# Python Modules and Packages

Modular programming allows you to break large programs into smaller, manageable files. This makes code reusable and organized.

## 1. Importing Built-in Modules
Python comes with many standard libraries.

````python
import math
import random

print(f"Square root of 25: {math.sqrt(25)}")
print(f"Random number between 1 and 100: {random.randint(1, 100)}")
````

## 2. Importing Specific Items and Aliasing
You can import exactly what you need to keep your namespace clean.

````python
from datetime import datetime as dt

current_time = dt.now()
print(f"Current Date and Time: {current_time}")
````

## 3. The `__name__ == '__main__'` Pattern
This pattern is used to check if a module is being run directly or being imported.

````python
def my_tool():
    print("Tool is performing work!")

if __name__ == "__main__":
    # This only runs if you run THIS notebook cell directly.
    print("Script started successfully.")
    my_tool()
else:
    print("I have been imported!")
````

## 4. Understanding Packages
A package is just a folder containing a file named `__init__.py`. 

**Example Structure:**
```
library/
  ├── __init__.py
  ├── books.py
  └── members.py
```
