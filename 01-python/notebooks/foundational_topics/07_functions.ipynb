# Python Functions and Recursion

Functions are reusable blocks of code. They can take inputs (parameters) and return outputs.

## 1. Basic Function Definition
Defining a function with `def` and calling it.

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Himanshu Nigam"))
```

## 2. Argument Types
- **Default Parameters**: Used when an argument is omitted.
- **Keyword Arguments**: Providing arguments by name.

```python
def power(base, exponent=2): # Default value for exponent
    return base ** exponent

print(f"3 squared: {power(3)}")            # Uses default exponent=2
print(f"2 cubed: {power(exponent=3, base=2)}") # Using keyword arguments
```

## 3. Variable-Length Arguments (*args and **kwargs)
Sometimes you don't know in advance how many arguments will be passed into a function.

### `*args` (Non-Keyword Arguments)
- **Collector**: It collects extra positional arguments into a **tuple**.
- **Use Case**: Functions that handle a dynamic number of items, like a calculator's `sum` or a logger.

```python
def sum_all(*numbers):
    print(f"Processing a tuple of {len(numbers)} numbers: {numbers}")
    return sum(numbers)

print(f"Total: {sum_all(10, 20, 30, 40)})")
```

### `**kwargs` (Keyword Arguments)
- **Collector**: It collects extra keyword arguments into a **dictionary**.
- **Use Case**: Handling configurations, user profiles, or any scenario where named options are dynamic.

```python
def show_profile(name, **details):
    print(f"User: {name}")
    print(f"Details: {details}")

show_profile("Alice", number_of_rows=3, role="Engineer")
```

### Professional Use Case: Wrappers & Forwarding
The most powerful use of `*args` and `**kwargs` is for **forwarding** arguments to other functions, which is the foundation of Python Decorators.

```python
def logger_wrapper(func, *args, **kwargs):
    print(f"Logging: Calling function '{func.__name__}'")
    return func(*args, **kwargs)

result = logger_wrapper(power, 5, exponent=3)
print(f"Result: {result}")
```

## 4. Lambda Functions
Anonymous functions for simple operations. They are often used with higher-order functions.

```python
multiply = lambda x, y: x * y
print(f"5 * 6 = {multiply(5, 6)}")
```

## 5. Higher-Order Functions
Functions that take other functions as arguments. Common examples include `map()` and `filter()`.

```python
numbers = [1, 2, 3, 4, 5, 6]

# map: Square all numbers
squares = list(map(lambda x: x**2, numbers))
print(f"Squares: {squares}")

# filter: Keep only even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(f"Evens: {evens}")
```

## 6. Recursion
A function calling itself.

### Importance of Base Case
Every recursive function needs a **base case** to stop it from calling itself forever. Without it, you get a "RecursionError" (infinite loop).

### Avoiding Infinite Recursion
1. **Base Case**: Handle the simplest version of the problem.
2. **Progress**: Each call must move closer to the base case.

```python
def factorial(n):
    # Base case: if n is 1 or 0, return 1
    if n <= 1:
        return 1
    # Recursive step: call factorial with n-1
    return n * factorial(n - 1)

print(f"Factorial of 20 is: {factorial(20)}")
```
