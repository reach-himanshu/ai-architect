# Python Variable Scope (LEGB)

Scope defines where a variable can be accessed. Python follows the **LEGB** rule: **L**ocal, **E**nclosing, **G**lobal, **B**uilt-in.

## 1. Local vs Global Scope

```python
x = "Global X"  # Global variable

def demo():
    y = "Local Y" # Local variable
    print(f"Inside function: {x}")
    print(f"Inside function: {y}")

demo()
print(f"Outside function: {x}")
# print(y) # This would fail
```

## 2. Modifying Global Variables
Use the `global` keyword to change a global variable from inside a function.

```python
counter = 0

def increment():
    global counter
    counter += 1
    print(f"Counter inside: {counter}")

increment()
increment()
print(f"Counter outside: {counter}")
```

## 3. Enclosing Scope and `nonlocal`

```python
def outer():
    text = "Original Outer"
    
    def inner():
        nonlocal text
        text = "Modified by Inner"

    inner()
    print(f"Outer text: {text}")

outer()
```

## 4. Why Use Nested Functions? (Practical Example)

### The Closure Pattern
A **closure** is a function object that remembers values in enclosing scopes even if they are not present in memory. This is often used for **Factory Functions** or preserving state.

```python
def make_multiplier(n):
    # Enclosing scope: n is stored here
    def multiply(x):
        return x * n
    return multiply

times3 = make_multiplier(3)
times5 = make_multiplier(5)

print(f"10 multiplied by 3: {times3(10)}")
print(f"10 multiplied by 5: {times5(10)}")
```

### State Management
Using `nonlocal` to keep a running count without a global variable.

```python
def simple_counter():
    count = 0
    def increment():
        nonlocal count
        count += 1
        return count
    return increment

counter = simple_counter()
print(f"First call: {counter()}")
print(f"Second call: {counter()}")
```

1. What is Enclosing Scope? If you have a function inside another function, the code inside the inner function can see variables defined in the outer function. That "space" between them is the enclosing scope.

2. Why use nonlocal? Just like you need global to change a variable at the top level, you need nonlocal to change a variable that belongs to an outer function. If you don't use it, Python will just create a new local variable with the same name.

3. Why create functions within functions? Here are the three main reasons why we do this in professional Python:

Closures: To create functions that "remember" a piece of data.
Example in Notebook: Creating a multiply_by_3 and multiply_by_5 function from a single "multiplier factory" function.
Encapsulation: If a helper function is only needed for one specific task, hiding it inside the main function keeps your code cleaner and prevents other parts of the program from accidentally calling it.
Decorators: This is the most advanced use. It's how we wrap additional logic (like logging or security checks) around existing functions.
