# Python Generators (`yield`)

Generators allow you to declare a function that behaves like an iterator. They are memory efficient because they generate values on the fly instead of storing them in memory.

## 1. The `yield` Statement
When a generator function is called, it returns a generator object without even beginning execution of the function. When the `next()` method is called, the function starts executing until it reaches the `yield` keyword.

```python
def simple_generator():
    yield "First"
    yield "Second"
    yield "Third"

gen = simple_generator()
print(next(gen))
print(next(gen))
print(next(gen))
```

## 2. Memory Efficiency Comparison
Watch how a generator is much lighter than a list when dealing with large ranges.

```python
import sys

n = 10000
my_list = [x for x in range(n)]
my_gen = (x for x in range(n))

print(f"Size of list in memory: {sys.getsizeof(my_list)} bytes")
print(f"Size of generator in memory: {sys.getsizeof(my_gen)} bytes")
```

## 3. Practical Use: Fibonacci Sequence
Generators are perfect for mathematical sequences that could be very long.

```python
def fibonacci(limit):
    a, b = 0, 1
    while a < limit:
        yield a
        a, b = b, a + b

print("Fibonacci numbers up to 50:")
for num in fibonacci(5050):
    print(num)
```
