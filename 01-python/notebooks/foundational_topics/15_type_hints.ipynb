# Type Hints

Static typing for Python.

```python
from typing import List, Dict, Optional, Callable

# 1. BASIC TYPES
def calculate_total(prices: List[float], tax: float = 0.1) -> float:
    subtotal = sum(prices)
    return subtotal * (1 + tax)

print(calculate_total([10.0, 20.0, 30.0]))
```

```python
# 2. OPTIONAL (Value or None)
def find_user(user_id: int) -> Optional[Dict[str, str]]:
    users = {1: {"name": "Alice"}}
    return users.get(user_id)  # Returns None if not found

print(find_user(1))
print(find_user(99))
```

```python
# 3. CALLABLE (Function as argument)
def apply_operation(a: int, b: int, op: Callable[[int, int], int]) -> int:
    return op(a, b)

result = apply_operation(10, 5, lambda x, y: x * y)
print(f"Result: {result}")
```
