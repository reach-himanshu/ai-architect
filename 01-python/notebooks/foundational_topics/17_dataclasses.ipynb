# Dataclasses & Named Tuples

Reducing boilerplate in data classes.

```python
from dataclasses import dataclass, field

@dataclass
class User:
    name: str
    email: str
    age: int = 0
    tags: list = field(default_factory=list)

u1 = User("Alice", "alice@example.com", 30)
u2 = User("Alice", "alice@example.com", 30)

print(u1)
print(f"Equal? {u1 == u2}")  # True - auto __eq__
```

```python
from typing import NamedTuple

class APIResponse(NamedTuple):
    status: int
    body: str

resp = APIResponse(200, '{"data": []}')

# Tuple unpacking works!
status, body = resp
print(f"Status: {status}")
```
