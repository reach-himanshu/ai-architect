# Python File I/O (Reading and Writing Files)

Interacting with the file system is a core part of most Python applications. We use the `open()` function, preferably within a `with` block.

## 1. Writing to a File
We'll start by creating a sample file.

```python
content = """Line 1: Hello World
Line 2: Python is amazing
Line 3: File I/O is easy"""

with open("example.txt", "w") as f:
    f.write(content)
print("File created successfully.")
```

## 2. Reading from a File
There are several ways to read content.

```python
print("--- Iterating line by line (Efficient) ---")
with open("example.txt", "r") as f:
    for line in f:
        print(line.strip())

print("\n--- Reading everything as a list ---")
with open("example.txt", "r") as f:
    lines = f.readlines()
    print(lines)
```

## 3. Appending to a File
Use mode `'a'` to add content without deleting what's already there.

```python
with open("example.txt", "a") as f:
    f.write("\nLine 4: This was appended!")

with open("example.txt", "r") as f:
    print(f.read())
```

## 4. Path Handling with `pathlib`

```python
from pathlib import Path

file_path = Path("example.txt")
if file_path.exists():
    print(f"File '{file_path.name}' exists and its size is {file_path.stat().st_size} bytes.")
```
