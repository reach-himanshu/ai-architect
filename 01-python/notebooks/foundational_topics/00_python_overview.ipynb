````python
class LoanApplication: # this is a class definition
    # this is a constructor
    def __init__(self, user_name, loan_amount):
        self.loan_amount = loan_amount
        self.user_name = user_name
# this method displays the loan application details
    def show_loan_app_details(self):
        print(f"Loan Applicant Name: {self.user_name}")
        print(f"Loan Amount: {self.loan_amount}")

        
````

````python
new_loan = LoanApplication('Sam Altman', 2000000.00)
new_loan.show_loan_app_details()
````

While methods like __init__ define behavior, dunders like __name__, __doc__, and __annotations__ are metadata attributes. They store information about your code rather than defining how it runs.

1. __name__ (The Scope Identifier)
This is the most famous dunder attribute. It tells you the name of the "namespace" where the code is currently running.

If you run a file directly: Python sets __name__ to "__main__".
If you import a file: Python sets __name__ to the filename (e.g., "math_utils").
Common Pattern:

python
def main():
    print("Doing work...")
if __name__ == "__main__":
    # This block only runs if you execute THIS file.
    # It won't run if someone else imports this file.
    main()
2. __doc__ (The Documentation String)
This stores the "Docstring" you write at the very top of a function, class, or module. Tools like pydoc or IDEs use this to show you help text.

python
def add(a, b):
    """Adds two numbers and returns the result."""
    return a + b
print(add.__doc__) 
# Output: Adds two numbers and returns the result.
3. __annotations__ (Type Metadata)
When you use Type Hints in Python, they are stored in this dictionary. This is used by static type checkers (like Mypy) or for runtime validation.

python
def greet(name: str, age: int) -> str:
    return f"Hello {name}, you are {age}"
print(greet.__annotations__)
# Output: {'name': <class 'str'>, 'age': <class 'int'>, 'return': <class 'str'>}
4. Other Important Dunder Attributes
Attribute	Purpose
__file__	Returns the absolute path to the current script file.
__dict__	A dictionary containing all the attributes of an object (its "state").
__module__	The name of the module in which the object was defined.
__all__	A list of strings defining what gets exported when someone does from module import *.
Summary Comparison
Dunder Methods (__init__, __str__): Control how objects act.
Dunder Attributes (__name__, __doc__): Describe what objects are.

# Python Standard Data Types

Python has several built-in data types that are used to store and manipulate data. They are categorized as follows:

## 1. Numeric Types
Used to represent numbers.
- **`int`**: Integer values (e.g., `5`, `-10`).
- **`float`**: Floating-point numbers (e.g., `3.14`, `-0.001`).
- **`complex`**: Complex numbers with a real and imaginary part (e.g., `3 + 5j`).

## 2. Sequence Types
Used to store collections of items in a specific order.
- **`str`**: Strings; sequences of Unicode characters (e.g., `"Hello"`).
- **`list`**: Mutable sequences (e.g., `[1, "apple", 3.5]`).
- **`tuple`**: Immutable sequences (e.g., `(10, 20, 30)`).
- **`range`**: Represents an immutable sequence of numbers (e.g., `range(5)`).

## 3. Mapping Type
Used to store data in key-value pairs.
- **`dict`**: Dictionaries (e.g., `{"name": "Alice", "age": 25}`).

## 4. Set Types
Used to store unordered collections of unique items.
- **`set`**: Mutable set (e.g., `{1, 2, 3}`).
- **`frozenset`**: Immutable version of a set.

## 5. Boolean Type
Represents truth values.
- **`bool`**: `True` and `False`.

## 6. Binary Types
Used to manipulate binary data.
- **`bytes`**: Immutable sequences of single bytes.
- **`bytearray`**: Mutable version of `bytes`.
- **`memoryview`**: Allows accessing internal data of an object without copying.

## 7. None Type
- **`NoneType`**: Represents the absence of a value (`None`).

### Type Checking
You can check the type of any object using the `type()` function:
```python
x = 5
print(type(x)) # Output: <class 'int'>
```


````python
temperature = int(101.2)
print(temperature)
print(type(temperature))
 
temperature = float(101.2)
print(temperature)
print(type(temperature))
 
temperature = bool(101.2)
print(temperature)
print(type(temperature))
 
temperature = str(101.2)
print(temperature)
print(type(temperature))
````
