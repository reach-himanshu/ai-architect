# Python OOP: Inheritance and Polymorphism

These concepts allow you to build complex systems by reusing code and defining common interfaces for different types of objects.

## 1. Inheritance
A child class can inherit all methods and attributes from a parent class.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return "Generic sound"

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

d = Dog("Buddy")
c = Cat("Mittens")

print(f"{d.name} says: {d.speak()}")
print(f"{c.name} says: {c.speak()}")
```

## 2. Using `super()`
The `super()` function allows you to call the method of the parent class. It is essential when you want to extend the behavior of the parent's `__init__` method.

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

class Developer(Employee):
    def __init__(self, name, salary, language):
        super().__init__(name, salary)
        self.language = language

dev = Developer("Alice", 90000, "Python")
print(f"{dev.name} codes in {dev.language}.")
```

## 3. Polymorphism
Polymorphism allows different objects to be accessed through the same interface (same method name) but perform different actions.

```python
class Square:
    def __init__(self, side):
        self.side = side
    def area(self):
        return self.side * self.side

class Circle:
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14 * self.radius * self.radius

shapes = [Square(4), Circle(5), Square(2)]

for shape in shapes:
    print(f"Area: {shape.area()}")
```
