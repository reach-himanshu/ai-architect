# Python OOP: Composition vs. Inheritance

Understanding when to use "Is-a" vs. "Has-a" relationships is key to writing professional, maintainable code.

## 1. The Inheritance Trap
Sometimes, it's tempting to use inheritance just to share code, even if the logic doesn't fit.

```python
class User:
    def __init__(self, name):
        self.name = name

class Admin(User):
    def delete_db(self):
        return f"{self.name} deleted the database!"

a = Admin("Himanshu")
print(a.delete_db())
```

## 2. Using Composition for Flexibility
Instead of making everything a subclass, give classes components.

```python
class Engine:
    def start(self): return "Engine Vroom!"

class Radio:
    def play(self): return "Playing music..."

class Car:
    def __init__(self):
        self.engine = Engine() # Composition
        self.radio = Radio()   # Composition

my_car = Car()
print(my_car.engine.start())
print(my_car.radio.play())
```
