# Python OOP: Special (Dunder) Methods

Dunder methods allow your objects to interact with built-in Python functions and operators. This is called **Operator Overloading**.

## 1. String Representation
Control how your object looks when printed.

```python
class Account:
    def __init__(self, owner, amount):
        self.owner = owner
        self.amount = amount

    def __str__(self):
        return f"Account owner: {self.owner} | Balance: {self.amount}"

    def __repr__(self):
        return f"Account('{self.owner}', {self.amount})"

acc = Account("Himanshu", 1000)
print(acc)         # Calls __str__
print(repr(acc))   # Calls __repr__
```

## 2. Arithmetic Overloading
Let's create a `Currency` class that can be added together.

```python
class Currency:
    def __init__(self, value, unit="USD"):
        self.value = value
        self.unit = unit

    def __add__(self, other):
        if self.unit != other.unit:
            raise ValueError("Cannot add different currencies!")
        return Currency(self.value + other.value, self.unit)

    def __repr__(self):
        return f"{self.value} {self.unit}"

c1 = Currency(50)
c2 = Currency(30)
print(f"Total: {c1 + c2}")
```

## 3. Length and Indexing
Making an object behave like a container.

```python
class Team:
    def __init__(self, members):
        self.members = members

    def __len__(self):
        return len(self.members)

    def __getitem__(self, index):
        return self.members[index]

t = Team(["Alice", "Bob", "Charlie"])
print(f"Team Size: {len(t)}")
print(f"Captain: {t[0]}")
```
