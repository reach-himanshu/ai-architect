# Python OOP: Mixins and Protocols

These patterns provide modern, decoupled ways to add functionality to your classes.

## 1. Using Mixins
Mixins are like "plugins" for your classes.

```python
class LoggerMixin:
    def log(self, message):
        print(f"[LOG]: {message}")

class Tool(LoggerMixin):
    def work(self):
        self.log("Tool is starting work.")

t = Tool()
t.work()
```

## 2. Protocols (Duck Typing with Type Hints)
Notice how we can use different classes without inheritance.

```python
from typing import Protocol

class Flyer_Protocol(Protocol):
    def fly(self) -> str:
        ...

class Bird:
    def fly(self): return "Flap flap"

class Airplane:
    def fly(self): return "Engines roaring"

def takeoff(entity: Flyer_Protocol):
    print(entity.fly())

takeoff(Bird())
takeoff(Airplane())
```
