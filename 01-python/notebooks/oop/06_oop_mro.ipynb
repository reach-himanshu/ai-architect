# Python OOP: Multiple Inheritance and MRO

Multiple inheritance allows a class to derive from multiple parents. Method Resolution Order (MRO) is the logic Python uses to decide which parent's method to call.

## 1. Multiple Inheritance Example
Notice how the order in the class definition affects the result.

```python
class Walker:
    def action(self):
        return "Walking on land"

class Swimmer:
    def action(self):
        return "Swimming in water"

class Amphibian(Swimmer, Walker):
    pass

frog = Amphibian()
print(f"Frog action: {frog.action()}")
```

## 2. Inspecting the MRO
You can see the exact path Python takes using `.mro()`.

```python
print("MRO for Amphibian:")
for cls in Amphibian.mro():
    print(cls)
```

## 3. The Diamond Problem and `super()`
Watch how `super()` visits classes based on the MRO order, not just the immediate parent.

```python
class Root:
    def show(self):
        print("Root")

class Left(Root):
    def show(self):
        print("Left")
        super().show()

class Right(Root):
    def show(self):
        print("Right")
        super().show()

class Child(Left, Right):
    def show(self):
        print("Child")
        super().show()

obj = Child()
obj.show()
```
