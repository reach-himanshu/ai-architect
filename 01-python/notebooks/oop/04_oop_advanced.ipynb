# Python OOP: Advanced Concepts

In this session, we'll explore `@classmethod`, `@staticmethod`, and the `@property` decorator.

## 1. Class Methods vs Static Methods
- **Class Methods** get access to the class via `cls`. Use them for Factory methods.
- **Static Methods** are just utility functions grouped inside the class namespace.

```python
class Pizza:
    def __init__(self, toppings):
        self.toppings = toppings

    def __repr__(self):
        return f"Pizza({self.toppings})"

    @classmethod
    def margherita(cls):
        return cls(["mozzarella", "tomatoes"])

    @classmethod
    def prosciutto(cls):
        return cls(["mozzarella", "tomatoes", "ham"])

    @staticmethod
    def validate_topping(topping):
        if topping == "pineapple": # The classic debate
            raise ValueError("No pineapple allowed!")
        return True

p1 = Pizza.margherita()
print(f"Menu Item 1: {p1}")

print(f"Validating topping: {Pizza.validate_topping('mushroom')}")
```

## 2. Using `@property` for Getters and Setters
Properties allow you to add logic around attribute access without breaking your API.

```python
class Celsius:
    def __init__(self, temperature=0):
        self._temperature = temperature

    def to_fahrenheit(self):
        return (self.temperature * 1.8) + 32

    @property
    def temperature(self):
        print("Getting value...")
        return self._temperature

    @temperature.setter
    def temperature(self, value):
        print("Setting value...")
        if value < -273.15:
            raise ValueError("Temperature below -273.15 is not possible")
        self._temperature = value

human_temp = Celsius(37)
print(f"Temp in F: {human_temp.to_fahrenheit()}")

human_temp.temperature = 40
print(f"New Temp: {human_temp.temperature}")
```
