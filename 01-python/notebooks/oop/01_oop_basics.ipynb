# Python Object-Oriented Programming (OOP)

OOP allows you to structure your code into reusable blueprints called **Classes**. From these classes, you create specific **Objects**.

## 1. Basic Class Definition
We'll define a simple `Robot` class.

```python
class Robot:
    def __init__(self, name, version):
        self.name = name
        self.version = version

    def say_hello(self):
        return f"Hello! I am {self.name}, version {self.version}."

r1 = Robot("Alpha", 1.0)
print(r1.say_hello())
```

## 2. Constructors and Destructors

### The Constructor: `__init__`
Automatically called when an object is created. Used to initialize state.

### The Destructor: `__del__`
Automatically called when an object is about to be deleted (garbage collected). Used for cleanup.

```python
class DatabaseConnection:
    def __init__(self, db_name):
        self.db_name = db_name
        print(f"--- Constructor: Connected to {self.db_name} ---")

    def __del__(self):
        print(f"--- Destructor: Connection to {self.db_name} closed safely ---")

print("Starting session...")
db = DatabaseConnection("User_DB")

print("Ending session manually...")
del db  # This triggers __del__ 
print("Session concluded.")
```

## 3. Class vs Instance Attributes
- **Instance Attributes** are unique. 
- **Class Attributes** are shared.

```python
class Robot:
    population = 0  # Class attribute

    def __init__(self, name):
        self.name = name
        Robot.population += 1
        print(f"Robot '{self.name}' created.")

print(f"Initial Population: {Robot.population}")
r1 = Robot("R2-D2")
r2 = Robot("C-3PO")
print(f"Final Population: {Robot.population}")
```

## 4. Methods acting on data

```python
class Circle:
    pi = 3.14159

    def __init__(self, radius=1):
        self.radius = radius

    def area(self):
        return self.radius * self.radius * Circle.pi

c = Circle(5)
print(f"Area of circle with radius 5: {c.area()}")
```
