# Packaging & Distribution

Creating reusable Python packages.

## Minimal pyproject.toml

```toml
[build-system]
requires = ["setuptools>=61.0"]
build-backend = "setuptools.build_meta"

[project]
name = "my-awesome-lib"
version = "0.1.0"
description = "Does awesome things"
dependencies = []
```

## Commands

```bash
# Install in editable/dev mode
pip install -e .

# Build distribution
python -m build

# Upload to PyPI
twine upload dist/*
```
