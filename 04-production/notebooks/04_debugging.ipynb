# Debugging & Profiling

Finding and fixing performance issues.

```python
import cProfile
import pstats

def slow_function():
    total = 0
    for i in range(100000):
        total += i
    return total

def fast_function():
    return sum(range(100000))

# Profile the functions
print("Profiling slow_function:")
cProfile.run('slow_function()', sort='cumtime')

print("\nProfileing fast_function:")
cProfile.run('fast_function()', sort='cumtime')
```
