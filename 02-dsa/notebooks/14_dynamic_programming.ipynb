# DSA 14: Dynamic Programming

Optimization using Memoization and Tabulation.

## 1. Fibonacci (Memoized)
Top-down approach.

```python
memo = {}

def fib_memo(n):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    
    memo[n] = fib_memo(n-1) + fib_memo(n-2)
    return memo[n]

print(f"Fibonacci(50): {fib_memo(50)}")
```

## 2. Climbing Stairs (Tabulation)
Count ways to reach the top. You can take 1 or 2 steps.
Bottom-up approach.

```python
def climb_stairs(n):
    if n <= 2:
        return n
    
    dp = [0] * (n + 1)
    dp[1] = 1
    dp[2] = 2
    
    for i in range(3, n + 1):
        dp[i] = dp[i-1] + dp[i-2]
        
    return dp[n]

print(f"Ways to climb 5 stairs: {climb_stairs(5)}")
```
