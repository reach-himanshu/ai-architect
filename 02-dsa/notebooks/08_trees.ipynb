# DSA 08: Trees

Implementing a Binary Search Tree (BST) and traversal algorithms.

```python
class TreeNode:
    def __init__(self, val=0):
        self.val = val
        self.left = None
        self.right = None

def insert_bst(root, val):
    if not root:
        return TreeNode(val)
    if val < root.val:
        root.left = insert_bst(root.left, val)
    else:
        root.right = insert_bst(root.right, val)
    return root

def inorder_traversal(root):
    if root:
        inorder_traversal(root.left)
        print(root.val, end=" ")
        inorder_traversal(root.right)

# Create Tree
root = TreeNode(10)
insert_bst(root, 5)
insert_bst(root, 15)
insert_bst(root, 3)
insert_bst(root, 7)

print("In-order Traversal (Sorted):")
inorder_traversal(root)
```
