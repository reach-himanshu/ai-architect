# DSA 06: Hash Tables

Using Hash Maps (Dictionaries) for efficient lookups.

## 1. Counting Frequencies
Problem: Count how many times each character appears in a string.

```python
text = "hello world"
freq = {}

for char in text:
    freq[char] = freq.get(char, 0) + 1

print(freq)
```

## 2. The "Two Sum" Problem
Given an array `nums` and a `target`, return indices of the two numbers such that they add up to `target`.
Target: 9, Nums: [2, 7, 11, 15] -> Returns [0, 1] because 2+7=9.

```python
def two_sum(nums, target):
    seen = {} # Value -> Index
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []

print(two_sum([2, 7, 11, 15], 9))
```
