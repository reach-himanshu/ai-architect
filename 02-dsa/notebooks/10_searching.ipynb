# DSA 10: Searching

Implementing Binary Search (Iterative and Recursive).

```python
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
            
    return -1 # Not found

sorted_nums = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
print(f"Found 23 at index: {binary_search(sorted_nums, 23)}")
print(f"Found 100 at index: {binary_search(sorted_nums, 100)}")
```
