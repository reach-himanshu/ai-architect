# DSA 04: Stacks

Solving the classic "Valid Parentheses" problem using a Stack.

```python
def is_valid_parentheses(s):
    stack = []
    mapping = {")": "(", "}": "{", "]": "["}

    for char in s:
        if char in mapping:
            # Closing bracket: Check top of stack
            top_element = stack.pop() if stack else '#'
            if mapping[char] != top_element:
                return False
        else:
            # Opening bracket: Push to stack
            stack.append(char)

    return not stack # True if stack is empty

print(f"()[]{}: {is_valid_parentheses('()[]{}')}")
print(f"([)]: {is_valid_parentheses('([)]')}")
```
