# DSA 13: Greedy Algorithms

Implementing a Greedy Coin Change solution.

```python
def greedy_coin_change(amount, coins):
    # Coins must be sorted descending for this greedy strategy to work on standard currency
    coins.sort(reverse=True)
    count = 0
    details = []
    
    for coin in coins:
        if amount == 0:
            break
        num_coins = amount // coin
        if num_coins > 0:
            count += num_coins
            amount -= num_coins * coin
            details.append(f"{num_coins}x{coin}")
            
    return count, details

amount = 87
coins = [25, 10, 5, 1]
total, breakdown = greedy_coin_change(amount, coins)
print(f"To make {amount}: {total} coins ({', '.join(breakdown)})")
```
