# Heaps & Priority Queues

Using Python's `heapq` module.

```python
import heapq

# 1. Basic Heap Operations
nums = [5, 3, 8, 1, 2]
heapq.heapify(nums)
print(f"Heapified: {nums}")

heapq.heappush(nums, 0)
print(f"After push(0): {nums}")

smallest = heapq.heappop(nums)
print(f"Popped: {smallest}, Remaining: {nums}")
```

```python
# 2. Top-K Largest
data = [10, 4, 3, 20, 15, 8, 25]
top_3 = heapq.nlargest(3, data)
print(f"Top 3 largest: {top_3}")

bottom_3 = heapq.nsmallest(3, data)
print(f"Top 3 smallest: {bottom_3}")
```
