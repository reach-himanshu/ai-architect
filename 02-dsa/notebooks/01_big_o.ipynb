# Big O Notation and Complexity Analysis

This notebook demonstrates how different algorithms perform as the input size grows.

## 1. Comparing O(n) vs O(n²)
Let's see the time difference as we increase the size of a list.

```python
import time

def linear_search(n):
    total = 0
    for i in range(n):
        total += i
    return total

def quadratic_search(n):
    total = 0
    for i in range(n):
        for j in range(n):
            total += (i + j)
    return total

n_small = 1000
n_large = 5000

start = time.time()
linear_search(n_large)
print(f"Linear O(n) for {n_large}: {time.time() - start:.6f} seconds")

start = time.time()
quadratic_search(n_small)
print(f"Quadratic O(n²) for {n_small}: {time.time() - start:.6f} seconds")

start = time.time()
quadratic_search(n_large)
print(f"Quadratic O(n²) for {n_large}: {time.time() - start:.6f} seconds")
```

## 2. Space Complexity Demo
Comparing memory usage when creating lists.

```python
import sys

def constant_space(n):
    total = 0 # Only one variable, O(1) space
    for i in range(n):
        total += i
    return total

def linear_space(n):
    lst = [i for i in range(n)] # Creating a list of size n, O(n) space
    return sum(lst)

n = 100000
print(f"Space consumed by constant_space: {sys.getsizeof(constant_space(n))} (result variable size)")
print(f"Space consumed by list in linear_space: {sys.getsizeof([i for i in range(n)])} bytes")
```
