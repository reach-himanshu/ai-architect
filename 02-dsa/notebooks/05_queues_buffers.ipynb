Implementing Queues and a Circular Buffer.

```python
from collections import deque

# Basic Queue
queue = deque()
queue.append("First")
queue.append("Second")
print(f"Dequeued: {queue.popleft()}") # First
print(f"Next: {queue[0]}")            # Second
```

## 2. Circular Buffer Implementation
A custom fixed-size queue that overwrites old data when full (if desired) or blocks.

```python
class CircularBuffer:
    def __init__(self, capacity):
        self.capacity = capacity
        self.buffer = [None] * capacity
        self.head = 0
        self.tail = 0
        self.size = 0

    def enqueue(self, item):
        if self.size == self.capacity:
            print("Buffer is Full!")
            return
        self.buffer[self.tail] = item
        self.tail = (self.tail + 1) % self.capacity # Wrap around
        self.size += 1

    def dequeue(self):
        if self.size == 0:
            return None
        item = self.buffer[self.head]
        self.head = (self.head + 1) % self.capacity # Wrap around
        self.size -= 1
        return item

cb = CircularBuffer(3)
cb.enqueue(1)
cb.enqueue(2)
cb.enqueue(3)
cb.enqueue(4) # Full
print(cb.dequeue()) # 1
cb.enqueue(4) # Success now
```
