# Trie (Prefix Tree)

Efficient prefix-based search.

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False

class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end = True

    def search(self, word):
        node = self._find_node(word)
        return node is not None and node.is_end

    def starts_with(self, prefix):
        return self._find_node(prefix) is not None

    def _find_node(self, s):
        node = self.root
        for char in s:
            if char not in node.children:
                return None
            node = node.children[char]
        return node

trie = Trie()
for word in ["apple", "app", "apply", "banana"]:
    trie.insert(word)

print(f"search('app'): {trie.search('app')}")
print(f"starts_with('app'): {trie.starts_with('app')}")
```
